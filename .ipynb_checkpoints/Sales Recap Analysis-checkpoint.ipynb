```python
def pie_chart(df, col, title):
    labels = [i for i in df[col].value_counts().index]
    df[col].value_counts().plot.pie(figsize=(8,8),
                                                 pctdistance=0.4,
                                                 startangle=90,
                                                 textprops={"color": "white"},
                                                 wedgeprops={"linewidth":0})
    plt.title(title, fontsize=20, pad=5)
    plt.ylabel("")
    box = plt.gca().get_position()
    plt.gca().set_position([box.x0, box.y0, box.width * 0.8, box.height])
    plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), prop={"size":12, 
                                                                 "weight":"ultralight"})
    plt.show();
```

```python
pie_chart(sales_df, "Promotion", "Jenis Promosi")
```

### Model Prediksi

```python
corr = sales_df.corr()

f, ax = plt.subplots(figsize=(35, 10))

mask = np.triu(np.ones_like(corr, dtype=bool))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```

```python
penjualan_df = pd.DataFrame()
penjualan_df['Jumlah Penjualan'] = sales_df['Sales In Date'].value_counts().sort_index()
penjualan_df
```

```python
penjualan_df.plot()
```

```python
from pandas import datetime
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(penjualan_df)
pyplot.show()
```

```python
train_df = penjualan_df[:int(len(penjualan_df)*0.9)]
test_df = penjualan_df[int(len(penjualan_df)*0.9):len(penjualan_df)]
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_df, order=(2,1,0))
model_fit = model.fit()
# summary of fit model
print(model_fit.summary())
```

```python
# line plot of residuals
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
# summary stats of residuals
print(residuals.describe())
```

```python
predict = model_fit.forecast(5)
predict = predict[1:]
predict = pd.DataFrame(predict)
predict
```

Hasil yang diperoleh belum cukup menggambarkan prediksi yang baik, perlu dilakukan proses analsis lebih lanjut. Seperti pemilihan parameter ARIMA yang lebih pas, atau menggunakan model matematis lainnya untuk proses prediksi.
